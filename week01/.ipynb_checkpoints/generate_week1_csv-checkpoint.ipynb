# Generate `week1_materials_sample.csv`

**Instructor use only** — run this notebook once to generate the sample dataset for the Week 1 lab.

**What it does:** Fetches ~80 binary oxide entries from the Materials Project API and saves them as a clean CSV file that students load in Part C of `week1_starter.ipynb`.

**After running:** Copy `week1_materials_sample.csv` into the `week01/` folder of your GitHub repository, then commit and push.

**Requirements:** Your `.env` file must be in the repository root with a valid `MP_API_KEY`.

---

```python
# Cell 1 — Imports and API key
from dotenv import load_dotenv
import os
import pandas as pd
import numpy as np
from mp_api.client import MPRester

load_dotenv()
API_KEY = os.getenv("MP_API_KEY")

if API_KEY:
    print(f"✅ API key loaded ({len(API_KEY)} characters)")
else:
    print("❌ API key not found — check your .env file")
```

```python
# Cell 2 — Fetch binary oxides from Materials Project
# This takes ~15-30 seconds

print("Fetching binary oxides from Materials Project...")

with MPRester(API_KEY) as mpr:
    docs = mpr.summary.search(
        chemsys="*-O",
        num_elements=(2, 2),
        fields=[
            "material_id",
            "formula_pretty",
            "elements",
            "band_gap",
            "formation_energy_per_atom",
            "energy_above_hull",
            "is_stable",
            "volume",
            "density",
            "nsites",
        ]
    )

print(f"✅ Retrieved {len(docs)} binary oxide entries")
```

```python
# Cell 3 — Convert to DataFrame and clean

rows = []
for d in docs:
    rows.append({
        "material_id":               d.material_id,
        "formula":                   d.formula_pretty,
        "elements":                  "-".join(sorted([str(e) for e in d.elements])),
        "band_gap":                  round(d.band_gap, 4)                     if d.band_gap                     is not None else None,
        "formation_energy_per_atom": round(d.formation_energy_per_atom, 4)    if d.formation_energy_per_atom    is not None else None,
        "energy_above_hull":         round(d.energy_above_hull, 4)            if d.energy_above_hull            is not None else None,
        "is_stable":                 d.is_stable,
        "volume_per_atom":           round(d.volume / d.nsites, 4)            if d.volume and d.nsites          else None,
        "density":                   round(d.density, 4)                      if d.density                      is not None else None,
        "nsites":                    d.nsites,
    })

df = pd.DataFrame(rows)

# Drop rows with missing critical values
df_clean = df.dropna(subset=["band_gap", "formation_energy_per_atom", "volume_per_atom"])

print(f"Entries before cleaning: {len(df)}")
print(f"Entries after cleaning:  {len(df_clean)}")
print(f"\nColumn dtypes:")
print(df_clean.dtypes)
```

```python
# Cell 4 — Create a balanced sample of 80 entries
# 20 metals (band_gap = 0) + 60 non-metals (band_gap > 0)

metals    = df_clean[df_clean["band_gap"] == 0]
nonmetals = df_clean[df_clean["band_gap"] >  0]

n_metals    = min(20, len(metals))
n_nonmetals = min(60, len(nonmetals))

sample_metals    = metals.sample(n=n_metals,    random_state=42)
sample_nonmetals = nonmetals.sample(n=n_nonmetals, random_state=42)

df_sample = pd.concat([sample_metals, sample_nonmetals]) \
              .sample(frac=1, random_state=42) \
              .reset_index(drop=True)

print(f"Sample size:  {len(df_sample)} entries")
print(f"  Metals (band_gap = 0):    {len(sample_metals)}")
print(f"  Non-metals (band_gap > 0): {len(sample_nonmetals)}")
print(f"\nBand gap statistics:")
print(df_sample["band_gap"].describe().round(3))
print(f"\nFirst 5 rows:")
df_sample.head()
```

```python
# Cell 5 — Save to CSV

output_path = "week1_materials_sample.csv"
df_sample.to_csv(output_path, index=False)

print(f"✅ Saved {len(df_sample)} entries to '{output_path}'")
print(f"\nNext steps:")
print(f"  1. Move '{output_path}' into your week01/ folder")
print(f"  2. In GitHub Desktop: commit and push")
print(f"  3. Students can now run Part C of week1_starter.ipynb")
```

```python
# Cell 6 — Verify the saved file loads correctly

df_check = pd.read_csv(output_path)

print(f"✅ File loads correctly")
print(f"   Shape: {df_check.shape}")
print(f"   Columns: {df_check.columns.tolist()}")
print(f"   Missing values: {df_check.isnull().sum().sum()}")
print(f"\nSample rows:")
df_check.sample(5, random_state=1)
```
